## Колода карт

```python
def labGenCards(deck_size=36, jokers=False):
    """
    Функция для создания колоды игральных карт.

    Параметры:
    deck_size (int): Количество карт в колоде. По умолчанию 36.
    jokers (bool): Наличие джокеров в колоде. По умолчанию False.

    Возвращает:
    deck (list): Список кортежей, представляющих колоду карт.
    """
    if not isinstance(deck_size, int) or deck_size not in [24, 36, 52]:
        raise ValueError("deck_size должен быть одним из следующих значений: 24, 36, 52")
    if not isinstance(jokers, bool):
        raise ValueError("jokers должен быть булевым значением")

    suits = ['черви', 'бубны', 'крести', 'пики']
    ranks = ['6', '7', '8', '9', '10', 'J', 'Q', 'K', 'A']

    if deck_size == 52:
        ranks = ['2', '3', '4', '5'] + ranks
    elif deck_size == 24:
        ranks = ranks[3:]

    deck = [(rank, suit) for suit in suits for rank in ranks]

    if jokers:
        deck.extend([('Joker', 'black'), ('Joker', 'red')])

    return deck

# Пример использования функции
print(labGenCards())
print(labGenCards(52, True))
print(labGenCards(36, True))
print(labGenCards(24, False))
```
