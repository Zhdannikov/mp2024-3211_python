# Вставка картинки в картинку 

```python
from PIL import Image
```

```python
image_block = Image.open('image_block.png')
image_block
```

```python
# Преобразовать изображение в миниатюру (с сохранением пропорции)
image_block.thumbnail((100, 100))
image_block
```

```python
im = Image.new(size=(700, 200), mode= 'RGB')

im.paste(image_block)
im.paste(image_block, box= (200,0))
im.paste(image_block, box= (300,0))
im.paste(image_block, box= (400,0))
im
```

```python
def draw_images(im, imPaste,
                xy = None, 
                image_count=3,
                image_cols= None,
                image_height=None):
    """
    Параметры:
    im - объект изображения PIL.Image,
    imPaste - Объект вставляемого изображения,
    xy - Координаты XY, 
    image_count - количество вставок изображения,
    image_cols - Количество колонок изображений,
    image_height - высота вставляемой картинки 
    """

    x1, y1 = (0, 0) if xy is None else xy
    im_colors = image_cols if image_cols is None else image_cols
    im_height = imPaste.height if image_height is None else image_height

    if image_count > 20:
        raise 'image_count - Количество должно быть меньше 20'
    
    imPasteNew = imPaste.copy()
    imPasteNew.thumbnail((imPaste.width, im_height))

    im_rows = image_count // image_cols + 1
    print(im_rows)

    for r in range(im_rows):
        for c in range(im_colors):
            if r * im_colors + c < image_count:
                x = x1 + imPasteNew.width * c
                y = y1 + imPasteNew.height * r
                im.paste(imPasteNew, box =(x,y))

im = Image.new(size=(700,300), mode='RGB')
draw_images(im,imPaste=image_block, image_count = 19, image_cols=7, image_height=50)
im
```
