```python
url_base = 'https://static-maps.yandex.ru/1.x'
url_params = dict()
url_params['l'] = 'map' # схема
url_params['ll'] = '37.591934,55.751356' # Москва
url_params['z'] = 12
print(url_params)
```

```python
import requests
from PIL import Image
```

```python
url_params['l'] = 'map' # схема
with requests.get(url=url_base, params=url_params, stream=True) as r:
    im = Image.open(r.raw)
print(im.format)
print(im.mode)   
print(im.size) 
im
```
