# CSV + cities.csv

```python
import csv
```

```python
file_cities_csv = "cities.csv"
```

```python
with open(file_cities_csv) as f:
    reader = csv.reader(f, delimiter=';')
    for n, line in enumerate(reader):
        if n>=5:
            break
        print(line)
```

```python
with open(file_cities_csv) as f:
        reader = csv.DictReader(f, delimiter=';')
        for line in reader:
                print(line)
```

```python
with open(file_cities_csv) as f:
        reader = csv.DictReader(f, delimiter=';')
        print(reader.fieldnames)
```

```python
with open(file_cities_csv) as f:
        reader = csv.DictReader(f, delimiter=';')
        for line in reader:
          print(line['Город'])
```

```python
with open(file_cities_csv) as f:
        reader = csv.DictReader(f, delimiter=';')
        for line in reader:
                if 'Крым' in line['Регион']:
                        print(line)
```

```python
temp_header = ''
temp_data = []

with open(file_cities_csv) as f:
        reader = csv.DictReader(f, delimiter=';')
        temp_header = reader.fieldnames
        for line in reader:
                if line['Регион'] == 'Башкортостан':
                        temp_data.append(line)

print(temp_header)
for line in temp_data:
        print(line.values())                        
```

```python
import os

TEMP_DIR = 'temp'
if not os.path.exists(TEMP_DIR):
    os.mkdir(TEMP_DIR)
```

```python
with open(os.path.join(TEMP_DIR, 'temp_cities_Башкортостан.csv'), 'w') as f:
    writer = csv.DictWriter(f, fieldnames=temp_header, lineterminator='\n', delimiter=';')
    writer.writeheader()
    writer.writerows(temp_data)
```
