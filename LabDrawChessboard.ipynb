```python
import numpy as np
import matplotlib.pyplot as plt

def labDrawChessboard(size=8, cell_color='binary', background_color='binary'):
    # Создаем доску
    board = np.zeros((size, size, 3))

    # Заполняем клетки
    for i in range(size):
        for j in range(size):
            if (i + j) % 2 == 0:
                board[i, j] = plt.get_cmap(cell_color)(0)[:3]
            else:
                board[i, j] = plt.get_cmap(background_color)(1)[:3]

    # Отображаем доску
    plt.imshow(board)

    # Добавляем буквы и числа по краям
    plt.xticks(np.arange(size), list('ABCDEFGH'))
    plt.yticks(np.arange(size), list('87654321'))

    plt.show()

# Пример использования функции
labDrawChessboard(8, 'binary', 'grey')
```
