```python
from PIL import Image, ImageDraw
import math

img = Image.new('RGB', (500, 500), 'white')
d = ImageDraw.Draw(img)

# Центр изображения
center_x, center_y = 250, 250

# Радиус шестиугольника
radius = 200

# Координаты шестиугольника
hexagon = [(center_x + radius * math.cos(math.radians(angle)), center_y + radius * math.sin(math.radians(angle))) for angle in range(0, 360, 60)]

d.line(hexagon + [hexagon[0]], fill='black', width=4)
img
```

```python
from PIL import Image, ImageDraw
from math import cos, sin, radians

count = 80
im = Image.new(size=(500,500), mode='RGB', color='black')
draw = ImageDraw.Draw(im)
angle = 360 / count
stripe_width = 1  # ширина полосы

for i in range(count):
    # Рисуем основные линии
    draw.line([(im.width/2, im.height/2), (im.width/2 + im.width/2 * cos(radians(i * angle)), im.height/2 + im.height/2 * sin(radians(i * angle)))], fill='white')

# Добавляем полосу по контуру круга
draw.arc([(stripe_width, stripe_width), (im.width - stripe_width, im.height - stripe_width)], 0, 360, fill='white')

im
```

```python
from PIL import Image, ImageDraw
from math import cos, sin, radians, sqrt

count = 80
im = Image.new(size=(500,500), mode='RGB', color='black')
draw = ImageDraw.Draw(im)
angle = 360 / count
diagonal = sqrt(im.width**2 + im.height**2) / 2

for i in range(count):
    draw.line([(im.width/2, im.height/2), (im.width/2 + diagonal * cos(radians(i * angle)), im.height/2 + diagonal * sin(radians(i * angle)))], fill='white')

# Добавляем полоски по бокам
border_width = 1  # ширина полоски
draw.rectangle([0, 0, im.width, border_width], fill='white')  # верхняя полоска
draw.rectangle([0, im.height - border_width, im.width, im.height], fill='white')  # нижняя полоска
draw.rectangle([0, 0, border_width, im.height], fill='white')  # левая полоска
draw.rectangle([im.width - border_width, 0, im.width, im.height], fill='white')  # правая полоска

im
```

```python
from PIL import Image, ImageDraw
from math import cos, sin, radians, sqrt

count = 500  # увеличиваем количество лучей
im = Image.new(size=(500,500), mode='RGB', color='grey')
draw = ImageDraw.Draw(im)
angle = 360 / count

# Координаты нового центра
new_center_x = 450
new_center_y = 100

for i in range(count):
    # Рисуем основные линии
    end_x = center_x + center_x * cos(radians(i * angle))
    end_y = center_y + center_y * sin(radians(i * angle))
    
    # Проверяем, что конечная точка линии находится внутри круга
    if sqrt((end_x - center_x)**2 + (end_y - center_y)**2) <= center_x:
        draw.line([(new_center_x, new_center_y), (end_x, end_y)], fill='white')

im
```

```python
from PIL import Image, ImageDraw

# Создаем новое изображение
im = Image.new(size=(500,500), mode='RGB', color='black')
draw = ImageDraw.Draw(im)

# Разделите верхнюю сторону квадрата на 10 частей
points = [(i * im.width // 10, 0) for i in range(10)]

# Рисуем линии от нижнего левого угла до каждой из 10 точек вверху
for point in points:
    draw.line([(0, im.height), point], fill='white')

# Рисуем линии от правого верхнего угла до каждой точки на левой стороне
for i in range(10):
    draw.line([(im.width, 0), (0, i * im.height // 10)], fill='white')

im
```
